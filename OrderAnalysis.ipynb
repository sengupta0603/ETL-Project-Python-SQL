```python
#read data from the file and handle value
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#fixing column name formats, renanming them
df.rename(columns={'Order Id': 'order_id', 'City': 'city'})
```

```python
#Changing All columns name in a single line 
#df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ', '_')
```

```python
df.head(5)
```

```python
#derive  a new column discount, Selling price and profit
#df['discount'] = df['list_price']*df['discount_percent']*.01
#df['sale_price'] = df['list_price']- df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
#Converting order_date data type to date as its current data type is object
df.dtypes
```

```python
#Converting order_date data type to date as its current data type is object
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
#Droping cost price and discount percent columns as we don't require it anymore
df.drop(columns=['list_price', 'cost_price','discount_percent'],inplace=True)
```
